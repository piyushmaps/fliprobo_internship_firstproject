2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
# importing the libraries
from bs4 import BeautifulSoup
import requests

url="https://www.bcci.tv/international/fixtures"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify()) # print the parsed data of html
```

```python
team=[]
```

```python
 for i in soup.find_all('span',attrs={'u-unskewed-text fixture__tournament-label u-truncated'}):
        team.append(i.text)
```

```python
team
```

```python
len(team)
```

```python
date=[]
```

```python
 for i in soup.find_all('span',attrs={'fixture__datetime tablet-only'}):
        date.append((i.text).replace("\n","").split()[0:3])
```

```python
date
```

```python
time=[]
```

```python
 for i in soup.find_all('span',attrs={'fixture__datetime tablet-only'}):
        time.append((i.text).replace("\n","").split()[3:])
```

```python
time
```

```python
len(time)
```

```python
place=[]
```

```python
 for i in soup.find_all('p',attrs={'fixture__additional-info'}):
        place.append((i.text).replace("\n",""))
```

```python
place
```

```python
len(place)
```

```python
df=pd.DataFrame({
    "team":team,
    "date":date,
    "time":time,
    "place":place
})
```

```python
df
```
